```python
import matplotlib.pyplot as plt 
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
```

```python
plt.plot(x,y)
plt.show()
```

```python
# the letter 'r' from "ro" sets the color to red, and the 'o' sets what kind of line to plot
plt.plot(x,y,"r3")
plt.show()
```

```python
# the letter 'r' from "ro" sets the color to red, and the 'o' sets what kind of line to plot
plt.plot(x,y,"g*")
plt.show()
```
